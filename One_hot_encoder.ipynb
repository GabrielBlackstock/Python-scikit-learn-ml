```python
# Dictionary with data
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000],
     'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'],
     'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
```

```python
# Put the dictionary data into a pandas dataframe
df = pd.DataFrame(data=d)
```

```python
# Look into first 5 rows / columns of df
df.head()
```

```python
# Return unique city values
df['city'].unique()
```

```python
# One hot encoder = Transfer categorical data into numbers
# Each unique category is represented by a seperate binary (0 or 1) features
# Red becomes [1,0,0]
```

```python
# Implement one hot encoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# Define one hot encoder
# Ignore unknown columns
# Sparse output to return a numpy array
# Output the data in a pandas dataframe
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output = False). set_output(transform="pandas")
```

```python
# Transform data
# Fit examines city column in the df to identify uynique values
# Transform converts city column into one-hot encoded format
ohetransform = ohe.fit_transform(df[['city']])
```

```python
# Print transformed dataframe
ohetransform
```

```python
# Print original dataframe
df
```

```python
# Concatenate transformed df with old df and drop city
# Axis refers to direction of operation
# Axis = 1 refers to columns
ohe_concat = pd.concat([df, ohetransform], axis=1).drop(columns=['city'])
```

```python
# Print final dataframe
ohe_concat
```
